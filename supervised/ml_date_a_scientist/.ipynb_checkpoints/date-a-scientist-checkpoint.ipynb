### Comments on initatial data inspection

+ The vast majority of this dataset is made up of categorical variables. This will require either one-hot encoding for regression models or perhaps the use of Naive Bayes models.
+ It would seem that age is a requiered field for users, as it is the only variable in the dataset to not have any null values. 
+ Other multiple choice fields seem to be combined to give a better idea of the user. For example, we can observe responses like 'working on college' alongside 'graduated from college'. This will make one-hot encoding more difficult than if users had to choose from e.g. 5 specific responses.
+ There are a great many null values throughout the dataset.

### Cleaning text data

```python
# replace special characters
df.replace(r"&amp;", "&", regex=True, inplace=True)
df.replace(r"&rsquo;", "'", regex=True, inplace=True)

# replace html
df.replace(r"<[^<>]*>", " ", regex=True, inplace=True)
df.replace(r"https?://", " ", regex=True, inplace=True)

#replace \n
df.replace(r"\n", " ", regex=True, inplace=True)
```

```python
# Merge values for 'offspring'

print(df.offspring.unique())

# Create a mapping dictionary to merge values into fewer categories
mapping = {
    "doesn't have kids, but might want them": "No Kids, Might Want",
    "doesn't want kids": "No Kids, Doesn't Want",
    "doesn't have kids, but wants them": "No Kids, Wants",
    "doesn't have kids": "No Kids, Doesn't Want",  # Assuming 'doesn't have kids' means they don't want them
    'wants kids': "No Kids, Wants",
    'has a kid': "Has Kids, Might Want More",  # Assuming they might want more
    'has kids': "Has Kids, Might Want More",  # Assuming they might want more
    "doesn't have kids, and doesn't want any": "No Kids, Doesn't Want",
    "has kids, but doesn't want more": "Has Kids, Doesn't Want More",
    "has a kid, but doesn't want more": "Has Kids, Doesn't Want More",
    'has a kid, and wants more': "Has Kids, Wants More",
    'has kids, and might want more': "Has Kids, Might Want More",
    'might want kids': "No Kids, Might Want",
    'has a kid, and might want more': "Has Kids, Might Want More",
    'has kids, and wants more': "Has Kids, Wants More"
}

# Apply the mapping to your column, and fill NaN with 'Unknown'
df['kids_status_cleaned'] = df['offspring'].replace(mapping).fillna('Unknown')
```

```python
# Merge values for 'pets'

print(df.pets.unique())

# Create a mapping dictionary to merge values into fewer categories
pets_mapping = {
    'likes dogs and likes cats': 'Likes Both',
    'has cats': 'Likes Cats',
    'likes cats': 'Likes Cats',
    'has dogs and likes cats': 'Likes Both',
    'likes dogs and has cats': 'Likes Both',
    'likes dogs and dislikes cats': 'Likes Dogs, Dislikes Cats',
    'has dogs': 'Likes Dogs',
    'has dogs and dislikes cats': 'Likes Dogs, Dislikes Cats',
    'likes dogs': 'Likes Dogs',
    'has dogs and has cats': 'Likes Both',
    'dislikes dogs and has cats': 'Dislikes Dogs, Likes Cats',
    'dislikes dogs and dislikes cats': 'Dislikes Both',
    'dislikes cats': 'Dislikes Cats',
    'dislikes dogs and likes cats': 'Dislikes Dogs, Likes Cats',
    'dislikes dogs': 'Dislikes Dogs'
}

# Apply the mapping to your 'pets' column, and fill NaN with 'Unknown'
df['pets_status_cleaned'] = df['pets'].replace(pets_mapping).fillna('Unknown')
```

```python
# create a list of the essay column names
# essay0 - essay9
essays = list(map(lambda i: 'essay{}'.format(i), range(0, 10)))

# replace NaN with spaces
df[essays] = df[essays].fillna('')

# combine essays into one column
df['bios'] = df[essays].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
```

```python
df['bios'][3]
```

```python
# create final dataframe.
df_cleaned = df

# clean the religion, sign, and diet
# drop the na's
# Put into new dataframe df_cleaned

df_cleaned['religion'] = df_cleaned['religion'].apply(lambda x: str(x).split()[0])
df_cleaned['sign'] = df_cleaned['sign'].apply(lambda x: str(x).split()[0])
df_cleaned['diet'] = df_cleaned['diet'].apply(lambda x: str(x).split()[-1])
df_cleaned.replace('nan', np.nan, inplace=True)
```

```python
df_cleaned.columns
```

```python
print(df_cleaned['sign'].value_counts())
```

```python
df_cleaned.head()
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# nltk.download('vader_lexicon') commented out because it has already been installed

sia = SentimentIntensityAnalyzer()
```

```python
# Iterate through each row in the bio column to get sentiment scores, add it to a new column, split it, and drop it. 
# This takes a while.

def get_sentiment(row):
    return sia.polarity_scores(row['bios'])

df_cleaned['sentiment'] = df_cleaned.apply(get_sentiment, axis=1)
df_cleaned[['neg', 'neu', 'pos', 'compound']] = df_cleaned['sentiment'].apply(pd.Series)
df_cleaned.drop("sentiment", axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_cleaned[['age', 'height']] = scaler.fit_transform(df_cleaned[['age', 'height']])
```

```python
# columns to keep in the dataframe.
cleaned_columns = ['age', 'sex', 'body_type', 'diet', 'drinks', 'drugs', 'education', 'height', 'job', 'offspring', 'orientation', 'religion', 'sign', 'smokes', 'compound']

# features we want to transform to determine if someone has pets.
dummies_features = ['sex', 'body_type', 'diet', 'drinks', 'drugs', 'education', 'job', 'offspring', 'orientation', 'religion', 'sign', 'smokes']
```

```python
# create dummy variables for the pet model in a new pet dataframe
sign_df = df_cleaned[cleaned_columns].dropna()

for feature in dummies_features[1:]:
    pets_df = pd.get_dummies(sign_df, columns=[feature], prefix=[feature], dtype=int)
```
